```
#gradient descent with two variables
from sympy import *

x = Symbol('x')
y = Symbol('y')
x0=-1
y0=1
lr=0.01
ep=220
fx=x**2+y**2+10
dx = diff(x**2+y**2+10,x)
dy = diff(x**2+y**2+10,y)
for i in range(1,ep+1):
  x1=(-(lr)*(dx.subs(x,x0)))
  y1=(-(lr)*(dy.subs(y,y0)))
  x1=x0+x1
  y1=y0+y1
  x0=x1
  y0=y1
  print(round(x0,1),round(y0,1))
fx=fx.subs(x,x0)
fx=fx.subs(y,y0)
print(round(fx,2))
```
